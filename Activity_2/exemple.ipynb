```python
# exemplo01.py
from particle import Particle
import matplotlib.pyplot as plt

# Criate particle
p = Particle(x=0, y=0, vx=10, vy=10, mass=1)

# Simulation parameters
dt = 0.01  # time step
fx, fy = 0, -9.8  # gravity force (m/s²)

times = [1]
xs = [p.x]
ys = [p.y]
vxs = [p.vx]
vys = [p.vy]

t = 0
while p.y >= 0:
    p.newton(fx, fy, dt)
    t += dt
    times.append(t)
    xs.append(p.x)
    ys.append(p.y)
    vxs.append(p.vx)
    vys.append(p.vy)

# Plot trajectory
plt.figure(figsize=(10, 5))
plt.plot(xs, ys)
plt.title('Particle Trajectory (Oblique Launch)')
plt.xlabel('Position x (m)')
plt.ylabel('Position y (m)')
plt.grid(True)
plt.show()

```
