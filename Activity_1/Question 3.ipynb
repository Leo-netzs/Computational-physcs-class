```python
import numpy as np
import matplotlib.pyplot as plt

# FunÃ§Ã£o e sua derivada
def f(x):
    return x**4 + x / 4

def df(x):
    return 4 * x**3 + 1/4

# Gradiente descendente
def gradient_descent(x0, lr=0.01, max_iter=10000, tol=1e-8):
    x = x0
    history = [x]
    for _ in range(max_iter):
        grad = df(x)
        new_x = x - lr * grad
        history.append(new_x)
        if abs(new_x - x) < tol:
            break
        x = new_x
    return x, f(x), history

# ParÃ¢metros
x0 = 0.5  # ponto inicial
lr = 0.01

# Executar
x_min, y_min, path = gradient_descent(x0, lr)
print(f"ðŸŒ¿ MÃ­nimo encontrado: x = {x_min:.6f}, f(x) = {y_min:.6f}")

# Visualizar
x_vals = np.linspace(-1, 1, 500)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label="f(x) = xâ´ + x/4")
plt.scatter(path, [f(x) for x in path], color='red', s=10, label="Caminho da descida")
plt.scatter(x_min, y_min, color='green', s=80, label="MÃ­nimo encontrado")
plt.title("Gradiente Descendente na funÃ§Ã£o f(x) = xâ´ + x/4")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
