```python
import numpy as np
import matplotlib.pyplot as plt

# Function and its derivative
def f(x):
    return x**4 + x / 4

def df(x):
    return 4 * x**3 + 1/4

# Gradient Descent
def gradient_descent(x0, lr=0.01, max_iter=1000, tol=1e-8):
    x = x0
    history = [x]
    for _ in range(max_iter):
        grad = df(x)
        new_x = x - lr * grad
        history.append(new_x)
        if abs(new_x - x) < tol:
            break
        x = new_x
    return x, f(x), history

# Parameters
x0 = 1  # Initial point
lr = 0.001

# Execute gradient descent
x_min, y_min, path = gradient_descent(x0, lr)
print(f"🌿 Mínimum found: x = {x_min:.6f}, f(x) = {y_min:.6f}")

# Visualization
x_vals = np.linspace(-1, 1, 500)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label="f(x) = x⁴ + x/4")
plt.scatter(path, [f(x) for x in path], color='red', s=10, label="Descent path")
plt.scatter(x_min, y_min, color='green', s=80, label="Mínimum found")
plt.title("Gradient Descent in the Function f(x) = x⁴ + x/4")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
