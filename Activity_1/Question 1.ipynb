```python
import numpy as np
import matplotlib.pyplot as plt

# Definição da função e do gradiente
f = lambda x: x**2 - 1
def gradient(x):
    return 2*x  # Derivada de x^2 - 1

# Parâmetros do gradiente descendente
initial_position = 1.05  # Posição inicial
iterations = 50  # Número de iterações

# Geração dos pontos para o gráfico
x = np.linspace(-1.25, 1.25 , 10000)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$f(x) = x^2 -1$', color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Simulação do gradiente descendente com passos variáveis
x_pos = [initial_position]
y_pos = [f(initial_position)]
learning_rate = 0.9  # Taxa de aprendizado inicial



def adaptive_step(i):
    return max(0.01, learning_rate * (0.9 ** i))  # Passo decrescente a cada iteração

for i in range(iterations):
    step = adaptive_step(i)
    x_new = x_pos[-1] - step * gradient(x_pos[-1])
    x_pos.append(x_new)
    y_pos.append(f(x_new))

plt.plot(x_pos, y_pos, 'o-', color='red', label='Gradiente Descendente com Passos Variáveis')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradiente Descendente Adaptativo para encontrar o mínimo de f(x)')
plt.legend()
plt.grid()
plt.show()
x_pos = [initial_position]
y_pos = [f(initial_position)]

for _ in range(iterations):
    x_new = x_pos[-1] - learning_rate * gradient(x_pos[-1])
    x_pos.append(x_new)
    y_pos.append(f(x_new))

plt.plot(x_pos, y_pos, 'o-', color='red', label='Gradiente Descendente')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradiente Descendente para encontrar o mínimo de f(x)')
plt.legend()
plt.grid()
plt.show()
```
