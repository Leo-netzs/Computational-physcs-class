```python
import numpy as np
import matplotlib.pyplot as plt

# Function and gradient definition
f = lambda x: x**2 - 1
def gradient(x):
    return 2*x  # Derivative of x^2 - 1

# Gradient Descent Parameters
initial_position = 1.05  # Initial point
iterations = 50  # Number of iterations

# Generarte data for plotting
x = np.linspace(-1.25, 1.25 , 10000)
y = f(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$f(x) = x^2 -1$', color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Simulation of gradient descent with variable steps
x_pos = [initial_position]
y_pos = [f(initial_position)]
learning_rate = 0.9  # Initial learning rate



def adaptive_step(i):
    return max(0.01, learning_rate * (0.9 ** i))  # Decreasing step at each iteration

for i in range(iterations):
    step = adaptive_step(i)
    x_new = x_pos[-1] - step * gradient(x_pos[-1])
    x_pos.append(x_new)
    y_pos.append(f(x_new))

plt.plot(x_pos, y_pos, 'o-', color='red', label='Gradient Descent with Variable Steps')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Adaptive Gradient Descent to find the minimum of f(x)')
plt.legend()
plt.grid()
plt.show()
x_pos = [initial_position]
y_pos = [f(initial_position)]

for _ in range(iterations):
    x_new = x_pos[-1] - learning_rate * gradient(x_pos[-1])
    x_pos.append(x_new)
    y_pos.append(f(x_new))

plt.plot(x_pos, y_pos, 'o-', color='red', label='Gradient Descent')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent to find the minimum of f(x)')
plt.legend()
plt.grid()
plt.show()
```
